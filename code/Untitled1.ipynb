```python
h = 0.6766
Phi_star = 1.6 * 10 ** (-2) * h ** 3
M_star = -19.7 + 5*np.log10(h)
alpha = -0.9
def Phi(M):
    return 0.4 * np.log(10) * Phi_star * 10 ** (0.4 * (alpha+1 ) * (M_star - M)) * np.exp(-10 ** (0.4 * (M_star - M)))
```

```python
Ms = np.linspace(-16, -22, 1000)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(6, 5))
ax.semilogy(Ms, Phi(Ms),label="$\\alpha=-0.9,\;M^{*}=-19.7+ 5 \log h$")
ax.invert_xaxis()
ax.set_ylim([1e-4, 0.01])
ax.set_xlim([-16, -22])
ax.set_xlabel("$M$")
ax.set_ylabel("$\Phi/\mathrm{Mpc^{-3}}$")
ax.legend()
plt.tight_layout()
fig.savefig("typical curve.pdf")
```
